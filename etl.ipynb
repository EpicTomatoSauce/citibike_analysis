```python
import pandas as pd
import json
import os
```

```python
# import jan-20 csv and interpret results
jan_20 = pd.read_csv('JC-202001-citibike-tripdata.csv')
jan_20.head()
```

```python
# repeat above for remaining months csvs
feb_20 = pd.read_csv('JC-202002-citibike-tripdata.csv')
mar_20 = pd.read_csv('JC-202003-citibike-tripdata.csv')
apr_20 = pd.read_csv('JC-202004-citibike-tripdata.csv')
may_20 = pd.read_csv('JC-202005-citibike-tripdata.csv')
jun_20 = pd.read_csv('JC-202006-citibike-tripdata.csv')
jul_20 = pd.read_csv('JC-202007-citibike-tripdata.csv')
aug_20 = pd.read_csv('JC-202008-citibike-tripdata.csv')
sep_20 = pd.read_csv('JC-202009-citibike-tripdata.csv')
oct_20 = pd.read_csv('JC-202010-citibike-tripdata.csv')
nov_20 = pd.read_csv('JC-202011-citibike-tripdata.csv')
dec_20 = pd.read_csv('JC-202012-citibike-tripdata.csv')
# join all dfs together using concat (https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html)
frames = [jan_20, feb_20, mar_20, apr_20, may_20, jun_20, jul_20, aug_20, sep_20, oct_20, nov_20, dec_20]
combined_20 = pd.concat(frames)
```

```python
combined_20.head()
```

```python
# # from documentation (https://www.citibikenyc.com/system-data) amend Gender number to string
# # cannot replace int value with str so convert int to str first (https://www.geeksforgeeks.org/fastest-way-to-convert-integers-to-strings-in-pandas-dataframe/)
combined_20["gender"] = combined_20["gender"].map(str)
combined_20["gender"].replace({"1":"Male", "2":"Female", "Zero":"Unknown"}, inplace = True)
combined_20.head()
```
